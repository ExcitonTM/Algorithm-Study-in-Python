```python
import numpy as np
import heapq
```

```python
num_list = np.loadtxt('Median.txt')
```

```python
low_heap = list()
high_heap = list()
low_heap_size = len(low_heap)
high_heap_size = len(high_heap)
```

```python
med_list = list()
cur_med = np.inf
for idx, n in enumerate(num_list):
    print(str(idx)+':'+str(n))  
    print('current med:'+str(cur_med))
    if n < cur_med:
        heapq.heappush(low_heap, -n)
        low_heap_size += 1
    else:
        heapq.heappush(high_heap, n)
        high_heap_size += 1
        
    if high_heap_size > low_heap_size + 1:
        high_min = heapq.heappop(high_heap)
        heapq.heappush(low_heap, -high_min)
        high_heap_size -= 1
        low_heap_size += 1
    elif high_heap_size < low_heap_size - 1:
        low_max = -heapq.heappop(low_heap)
        heapq.heappush(high_heap, low_max)
        low_heap_size -= 1
        high_heap_size += 1
        
    if low_heap_size >= high_heap_size:
        cur_med = -low_heap[0]
    else:
        cur_med = high_heap[0]
        
    med_list.append(cur_med)
    
    if idx > 10000:
        break
 
    #print(low_heap)
    #print(high_heap)
```
