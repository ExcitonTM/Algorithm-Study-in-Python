```python
import numpy as np
import random
import copy
```

```python
def load_graph_txt(file_name):
    graph = {}
    with open(file_name) as f:
        for line in f:
            line_split = line.split()
            line_list_int = [int(i) for i in line_split]
            graph[line_list_int[0]] = line_list_int[1:]
    return graph
```

```python
graph_original = load_graph_txt('kargerMinCut.txt')
```

```python
def graph_contraction(v_1, v_2, graph): 
    connection_2 = graph[v_2]
    
    for v_2_i in connection_2:
        graph[v_2_i].remove(v_2)
        graph[v_2_i].append(v_1)
        
    graph[v_1] = graph[v_1] + connection_2
    graph[v_1].remove(v_1)
    graph.pop(v_2, None)
    
    return
```

```python
graph_contract = copy.deepcopy(graph_original)
```

```python
n = len(graph_contract.keys())
for i in range(n-2):
    v_1 = random.choice(list(graph_contract.keys()))
    connection_1 = graph_contract[v_1]
    print(connection_1)
    v_2 = random.choice(list(connection_1))
    print(v_2)
    graph_contraction(v_1, v_2, graph_contract)
```
